```python
import math
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split

# generate data
n = 1
m = 1000
X = np.arange(m)
y = .2 + .5 * X + np.random.normal(0, 5, size=(m))

# split sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# show some stats
print('Stats of X_train:')
print(f'\tmean = {np.mean(X_train)}')
print(f'\tmedian = {np.median(X_train)}')
print()
print('Stats of y_train:')
print(f'\tmean = {np.mean(y_train)}')
print(f'\tmedian = {np.median(y_train)}')

# plot the data
plt.title('Training set')
plt.scatter(X, y)
plt.show()

plt.subplot(1, 2, 1)
plt.title('X training set')
plt.boxplot(X)

plt.subplot(1, 2, 2)
plt.title('y training set')
plt.boxplot(y)

plt.show()
```

```python
# learn the parameters

from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

# run linear regression
linearRegression = linear_model.LinearRegression()
linearRegression.fit(np.array(X_train).reshape(-1, 1), y_train)
y_predict = linearRegression.predict(np.array(X_test).reshape(-1, 1))

# see results
print('\nTraining results:')
print(f'Learned parameters:')
print(f'\tintercept = {linearRegression.intercept_}')
print(f'\tcoef = {linearRegression.coef_}')
print(f'MSE = {mean_squared_error(y_test, y_predict)}')
print(f'R^2 score = {r2_score(y_test, y_predict)}')

# plot the prediction
plt.title('Prediction on test set')
plt.scatter(X_test, y_test)
plt.plot(X_test, y_predict, 'r')
plt.show()
```
